```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
load_boston()
```

```python
df=load_boston()
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
print(df.DESCR)
```

```python
## indepenedent features and dependent features
X=dataset
y=df.target
```

```python
X
```

```python
y
```

```python
# train_test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
## standardizing the dataset
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train_scaled=sc.fit_transform(X_train)
```

```python
X_train_scaled
```

```python
## if you want to inverse this transformed into old data
# sc.inverse_transform(X_train_scaled)
```

```python
X_test_scaled=sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train_scaled,y_train)
```

```python
## cross validation 
from sklearn.model_selection import cross_val_score
```

```python
mse=cross_val_score(regressor,X_train_scaled,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
mse
```

```python
np.mean(mse)
```

```python
## prediction
reg_pred=regressor.predict(X_test_scaled)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.distplot(reg_pred - y_test,kde=True)
```

```python
from sklearn.metrics import r2_score
```

```python
score=round(r2_score(reg_pred,y_test)*100,2)
```
